```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/s/chopin/l/grad/shadaab/Documents/Journal_Twitter/data/data_sampled.csv')
```

```python
def generate_wordcloud_series(data_frame, save_file_name, stp_words):

    text = ' '.join(data_frame['polished_text'].astype(str))

    custom_stopwords = set(['amp'])
    custom_stopwords.update(stp_words)
    stopwords = set(STOPWORDS)
    stopwords.update(custom_stopwords)

    wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)
    print(wordcloud.words_.keys())
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.savefig('/s/chopin/l/grad/shadaab/Documents/Journal_Twitter/word_cloud/' + save_file_name + ".pdf", 
                bbox_inches='tight', format='pdf')
    plt.show()
```

```python
df_ind = df[df['indorg']==-1]
generate_wordcloud_series(df_ind, 'ind', ['COVID19','COVID'])
```

```python
df_org = df[df['indorg']==1]
generate_wordcloud_series(df_org, 'org', ['COVID19','COVID'])
```
