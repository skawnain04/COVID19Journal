```python
#1 #2 #3
import json

# Initialize responses list for focus, information types, and language type
responses_focus = []

for i in range(0, len(tweets), chunk_size):
    chunk = tweets[i:i + chunk_size]

    # Define a more explicit question asking for focus, information type, and language type
    question = """Here is a list of tweets I have collected. I want you to classify each tweet based on three criteria:

    1. Classify each tweet as either 'collective' (if it focuses on collective outcomes like community health or public safety) or 'personal' (if it focuses on personal or immediate concerns).

    2. Additionally, determine if the tweet is providing 'preventive/authoritative' information (e.g., vaccine availability, safety protocols) or if it focuses on 'personal experience/opinion'.

    3. Finally, explore whether the tweet employs 'formal' or 'informal' language.

    Please respond with only the three classifications (e.g., 'collective, preventive/authoritative, formal' or 'personal, personal experience/opinion, informal') for each tweet in the order they appear, and avoid including any numeric labels.
    """ + "\n\n".join(chunk)

    # Send the prompt to the AI model for response
    response = client.chat.completions.create(
        model="benchChat",
        messages=[
            {"role": "system", "content": "You are a helpful assistant and an expert in analyzing tweet content."},
            {"role": "user", "content": question},
        ]
    )

    # Collect the response from the model for all three classifications
    responses_focus.append(response.choices[0].message.content)

# Process the responses and structure them as JSON
response_data = []

# Combine all responses into one string and split by lines
responses_str = "\n".join(responses_focus).split("\n")

# Function to remove any leading numbers or unnecessary characters
def clean_classification(text):
    return text.lstrip("0123456789. ").strip()

# Match each tweet with its corresponding classifications
for tweet, classification in zip(tweets, responses_str):
    # Split the classification into focus, info type, and language type
    classifications = classification.strip().lower().split(",")

    # Ensure valid classification length (should have 3 parts: focus, information type, and language type)
    if len(classifications) == 3:
        focus = clean_classification(classifications[0].strip())  # either 'collective' or 'personal'
        info_type = clean_classification(classifications[1].strip())  # either 'preventive/authoritative' or 'personal experience/opinion'
        language_type = clean_classification(classifications[2].strip())  # either 'formal' or 'informal'
    else:
        # If the model response is incomplete or unclear, assign default values
        focus = "unknown"
        info_type = "unknown"
        language_type = "unknown"

    response_data.append({
        "tweet": tweet,
        "focus": focus,
        "information_type": info_type,
        "language_type": language_type
    })

# Convert the structured data to JSON format
response_json = json.dumps(response_data, indent=4)

# Save the JSON data to a file
with open('/content/drive/MyDrive/Colab Notebooks/Twitter/org_tweet_focus_classification.json', 'w') as json_file:
    json_file.write(response_json)

print("JSON data has been saved to 'tweet_focus_info_and_language_classification.json'")
```
